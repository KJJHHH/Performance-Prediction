```python
import numpy as np
import pandas as pd
import win32com.client 
from openpyxl import load_workbook

def read_excel(filename):
    xlapp = win32com.client.Dispatch('Excel.Application')
    basic_information = xlapp.Workbooks.Open(
        filename,False, True, None,
        Password = 'NCCUAI')

    data_list = list(xlapp.Worksheets(2).UsedRange())
    data = pd.DataFrame(data_list[1:], columns = data_list[0])
    basic_information.Close()
    return data
```

# CALCULATE MEAN

```python
# Login mean
login_mean = (login['LOGIN_CNT']
    .groupby(login['AGENT_ID']).mean())
login_mean.reset_index()
data['login_mean'] = login_mean

# Contact mean
contact_mean = (contact['CONTACT_CNT']
    .groupby(contact['AGENT_ID']).mean())
contact_mean.reset_index()
data['contact_mean'] = contact_mean

contact_custmean = (contact['CONTACT_CUST_CNT']
    .groupby(contact['AGENT_ID']).mean())
contact_custmean.reset_index()
data['contact_custmean'] = contact_custmean

# Contact mean
visit_mean = (visit['VISIT_CNT']
    .groupby(visit['AGENT_ID']).mean())
visit_mean.reset_index()
data['visit_mean'] = visit_mean

visit_custmean = (visit['CUSTOMER_VISIT_CNT']
    .groupby(visit['AGENT_ID']).mean())
visit_custmean.reset_index()
data['visit_custmean'] = visit_custmean

# Client&Memo mean
client_mean = (client['CUSTOMER_CNT']
    .groupby(client['AGENT_ID']).mean())
client_mean.reset_index()
data['client_mean'] = client_mean

memo_mean = (client['MEMO_LENGTH']
    .groupby(client['AGENT_ID']).mean())
memo_mean.reset_index()
data['memo_mean'] = memo_mean

# Query mean
qnr_mean = (query['QNR_CNT']
    .groupby(query['AGENT_ID']).mean())
qnr_mean.reset_index()
data['qnr_mean'] = qnr_mean

qnr_custmean = (client['CUSTOMER_CNT']
    .groupby(query['AGENT_ID']).mean())
qnr_custmean.reset_index()
data['qnr_custmean'] = qnr_custmean

# Query mean
consult_mean = (consult['CONSULT_CNT']
    .groupby(consult['AGENT_ID']).mean())
consult_mean.reset_index()
data['consult_mean'] = consult_mean

consult_custmean = (consult['CONSULT_CUST_CNT']
    .groupby(consult['AGENT_ID']).mean())
consult_custmean.reset_index()
data['consult_custmean'] = consult_custmean

# Life insurance mean
life_mean = (life_insurance['PREMIUM']
    .groupby(life_insurance['AGENT_ID']).mean())
life_mean.reset_index()
data['life_insurance'] = life_mean

# Property insurance mean
property_mean = (property_insurance['PROPERTY_INSURANCE_PERMIUM']
    .groupby(property_insurance['AGENT_ID']).mean())
property_mean.reset_index()
data['property_insurance'] = property_mean
```
